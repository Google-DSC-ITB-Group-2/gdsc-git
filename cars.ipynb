Apakah tahun pembuatan mobil berpengaruh terhadap transmission mobil?

```python
df_trans=df.groupby(["transmission","year_produced"])["model_name"].count().to_frame().reset_index()
df_trans.loc[df_trans["transmission"]=="automatic"]
```

```python
df_trans.loc[df_trans["transmission"]=="automatic"].plot(kind="bar",x="year_produced",y="model_name")
```

```python
df_trans2=df.groupby(["transmission","year_produced"])["model_name"].count().to_frame().reset_index()
df_trans2.loc[df_trans2["transmission"]=="mechanical"]
```

```python
df_trans2.loc[df_trans2["transmission"]=="mechanical"].plot(kind="bar",x="year_produced",y="model_name")
```

Diperoleh analisis bahwa mobil edisi tahun terbaru kebanyakan menggunakan transmisi automatic, sedangkan mobil edisi tahun lama kebanyakan menggunakan transmisi mechanical
