<h3>Step 2: Data Cleaning</h3>
<p>Let's drop the features that irrelevant for my problem</p>
<p>Those features are ['title', 'channel_title', 'publish_time', 'tags', 'views', 'likes', 'dislikes', 'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed', 'description'] </p>

```python
df_cleaned = df.drop(['title', 'channel_title', 'publish_time', 'tags', 'views','likes', 'dislikes',
                      'thumbnail_link', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed',
                      'description'],axis=1)
df_cleaned
```

<p>After that, let's see if those features have null values</p>

```python
df_cleaned.isnull().sum()
```

<p>Pretty Nice! We don't have any null values in those feature<p>
<p>Normally, if there's some null values inside those feature, we can drop the data or we filled with some relevant data</p>
<p>Next thing, i want to do is create new data frame that's include the duration of trending videos, where each data is unique</p>

```python
#Compute the total days a video has trending
trend_days = df_cleaned.groupby(['video_id'])['video_id'].agg(total_trend_days=len).reset_index()

#Remove all duplicated video occurence and keep the last date of that video
df_cleaned_re = df_cleaned.sort_values(["video_id", "trending_date"])
df_cleaned_re = df_cleaned.drop_duplicates(subset=['video_id'], keep='last')

#Merge the dataset with the total days
df_cleaned_re = pd.merge(df_cleaned_re, trend_days, on='video_id')

df_cleaned_re =  df_cleaned_re.sort_values(['total_trend_days'], ascending=[0])
df_cleaned_re
```

<p>Before that, Let's do simple statistics first</p>

```python
df_cleaned_re.describe()
```

```python
#From statistics above, maybe we can find sum of of total_trend_days for each category
df_category = df_cleaned_re.groupby(['category_id']).sum().sort_values(['total_trend_days'], ascending=[0])
df_category['total_trend_days']
```
